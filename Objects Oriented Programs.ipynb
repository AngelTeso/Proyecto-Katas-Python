```python
#################################
######### POO- PROGRAMACIÓN #####
######### ORIENTADA #############
######### A OBJETOS #############

#La Programación Orientada a Objetos (POO) es un paradigma de programación que utiliza "objetos" para representar datos y métodos. 
# En lugar de centrarse en funciones y procedimientos, la POO se basa en la encapsulación de datos y comportamientos dentro de objetos, 
# lo que facilita la organización y reutilización del código.

# Conceptos clave de la POO: CLASES, OBJETOS (ATRIBUTOS), MÉTODOS Y ENCAPSULACIÓN

# 1. Clases: Son plantillas o moldes que definen las propiedades (atributos) y comportamientos (métodos) de los objetos.
# 2. Objetos: Son instancias de clases que contienen datos específicos y pueden realizar acciones definidas por sus métodos.
# 3. Encapsulación: Es el principio de ocultar los detalles internos de un objeto y exponer solo lo necesario a través de una interfaz pública.

class Empleado:
    """
    Clase que representa a un empleado en una empresa.
    """
    def __init__(self, nombre, edad, posicion, salario):
        """
        Constructor de la clase Empleado.

        Args:
            nombre (str): El nombre del empleado.
            edad (int): La edad del empleado.
            posicion (str): La posición del empleado.
            salario (float): El salario del empleado.
        """
        # Creamos atributos y asignamos el valor de los parámetros del método constructor
        self.nombre = nombre
        self.edad = edad
        self.posicion = posicion
        self.salario = salario

# Instanciamos un nuevo objeto(creamos un nuevo empleado)
juan = Empleado("Juan", 30, "Gerente", 50000)
```

```python
# Mostramos por terminal los atributos de nuestro objeto juan
print(f"La posición de Juan es: {juan.posicion}")
print(f"Todos los atributos de Juan son: {juan.__dict__}")
```

```python
# Desarrollamos el método de Subir Salario

def subir_salario(self, porcentaje):
        """
        Método para aumentar el salario del empleado en un cierto porcentaje.

        Args:
            porcentaje (float): El porcentaje de aumento del salario.
        """
        # Calcula el nuevo salario aumentado por el porcentaje dado
        self.salario *= (1 + porcentaje / 100)
        # Redondea el salario a dos decimales
        self.salario = round(self.salario, 2)
        # Muestra un mensaje indicando que el salario ha subido y cuál es el nuevo salario
        print(f"¡El salario de {self.nombre} ha subido! Nuevo salario: {self.salario:.2f}")
```

```python
subir_salario(juan, 10)  # Aumenta el salario de Juan en un 10%
```

```python
class Producto:
    """
    Clase que representa un producto para la venta.
    """
    def __init__(self, nombre, precio, plataforma_venta, IVA):
        """
        Constructor de la clase Producto.
        Args:
            nombre (str): El nombre del producto.
            precio (float): El precio del producto.
            plataforma_venta (str): La plataforma de venta del producto.
        """
        self.nombre = nombre  
        self.precio = precio  
        self.plataforma_venta = plataforma_venta
        self.IVA = IVA

    def impuestos(self):
        """
        Método para calcular los impuestos sobre el precio del producto.

        Returns:
            float: La cantidad de impuestos a pagar.
        """
        return self.precio * self.IVA
```

```python
class Electrodomestico(Producto):
    """
    Clase que representa un electrodoméstico como un tipo de producto.
    """
    def __init__(self, nombre, precio, plataforma_venta, IVA, certificado_energetico):
        """
        Constructor de la clase Electrodomestico.
        Args:
            nombre (str): El nombre del electrodoméstico.
            precio (float): El precio del electrodoméstico.
            plataforma_venta (str): La plataforma de venta del electrodoméstico.
            marca (str): La marca del electrodoméstico.
            consumo_energetico (str): El nivel de consumo energético del electrodoméstico.
        """
        # Llama al constructor de la clase madre 
        super().__init__(nombre, precio, plataforma_venta, IVA)
        
        # Añadimos los argumentos específicos de la clase hija
        self. certificado_energetico = certificado_energetico

    def evaluar_certificado(self):

        if self.certificado_energetico == "F":
            print(f"El producto {self.nombre} tiene un certificado energético deficiente")
```

```python
lavadora = Electrodomestico("lavadora siemens", 678, "online", 0.21, "F")
print(lavadora.nombre)
lavadora.evaluar_certificado()
# Output esperado:
#lavadora siemens
#El producto lavadora siemens tiene un certificado energético deficiente
```

```python
############################
###### POLIMORFISMO ########
############################

# EL POLIMORFISMO es un concepto fundamental en la Programación Orientada a Objetos (POO) que permite que objetos de diferentes clases sean tratados como objetos de una clase común. 
# Esto se logra mediante la definición de métodos con el mismo nombre en diferentes clases, lo que permite que una única interfaz pueda ser utilizada 
# para interactuar con objetos de distintas clases. 

class Producto:
        #... 
        def precio_iva(self):
                return self.precio * (1 + self.IVA)


class Electrodomestico(Producto):
        #... 
        def precio_iva(self):
                return self.precio * (1 + self.IVA) - 100
```

```python
lavadora = Electrodomestico("lavadora siemens", 678, "online", 0.21, "F")
```

```python
lavadora.precio_iva()  # Llama al método precio_iva de la clase Electrodomestico
```
